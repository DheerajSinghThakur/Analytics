**Problem 1. Import Titanic dataset and store as the pandas dataframe with name titanic**

**import library**

```python
import pandas  as pd
```

```python
titanic = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Titanic.csv')
```

```python
titanic.head()
```

**Problem 2. Create a new dataframe (female) from titanic of female passengers only and display first 8 rows**

```python
female = titanic[titanic['sex']=='female']
```

```python
female.head(8)
```

**Problem 3. Count number of female survived and not survived**

```python
female['survived'].value_counts()
```

**Problem 4. Count number of female survived and not survived as per passenger class**

```python
female.groupby(['survived','pclass']).count()
```

**Problem 5. Select passengers with age > 18 years**

```python
titanic.loc[titanic['age']>18]
```

**Problem 7. Select passengers with age > 18 years and female**

```python
titanic.loc[(titanic['age']>18) & (titanic['sex']=='female') ]
```

**Problem 8. Select passengers with age > 18 years and female and travelling by first class**

```python
titanic.loc[(titanic['age']>18) & (titanic['sex']=='female') & (titanic['pclass']==1), :]
```

**Problem 9. Select passengers with age > 18 years and female and travelling by first class and calculate their mean age**

```python
titanic.loc[(titanic['age']>18 ) & ( titanic['sex']=='female') & (titanic['pclass']==1 ), 'age'].mean()
```

**Problem 10. Select passengers with age > 18 years or female and travelling by first class and calculate their mean age**

```python
titanic.loc[(titanic['age']>18 ) | ( titanic['sex']=='female') & (titanic['pclass']==1 ), 'age'].mean()
```
